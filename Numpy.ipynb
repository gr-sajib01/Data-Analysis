```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("D:\\Python\\Python Project\\Data Analysis\\cleaned_Dataset.csv")
```

```python
print("Shape:", df.shape)
print("Columns:", df.columns)

print("First 5 rows:")
print(df.head())

print("Data Info:")
print(df.info())

print("Missing Values:")
print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print("Duplicate rows before removal:", duplicates)

before = df.shape[0]

df = df.drop_duplicates()

after = df.shape[0]

print("Row before:", before)
print("Row after:", after)
print("Duplicates removed:", before - after)

print("Remaining duplicates:", df.duplicated().sum())
```

```python
print("Missing values before cleaning:")
print(df.isnull().sum())

numeric_cols = df.select_dtypes(include=np.number).columns
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].mean())

df = df.dropna()
print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
print("Data types before:")
print(df.dtypes)

# Example fixes
df['over'] = df['over'].astype(int)
df['total_runs'] = df['total_runs'].astype(int)
df['is_wicket'] = df['is_wicket'].astype(int)

print("Data types after:")
print(df.dtypes)
```

```python
def phase(over):
    if over <= 6:
        return 'Powerplay'
    elif over <= 15:
        return 'Middle'
    else:
        return 'Last over'

df['phase'] = df['over'].apply(phase)

print(df[['over', 'phase']].head())
```

```python
def run_type(runs):
    if runs == 0:
        return 'Dot'
    elif runs <= 3:
        return 'Low'
    else:
        return 'High'

df['run_type'] = df['total_runs'].apply(run_type)

print(df[['total_runs', 'run_type']].head())
```

```python
phase_runs = df.groupby('phase')['total_runs'].sum()

print("Runs by phase:")
print(phase_runs)
```

```python
phase_avg = df.groupby('phase')['total_runs'].mean()

print("Average runs per ball by phase:")
print(phase_avg)
```
