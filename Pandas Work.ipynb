```python
def get_phase(over_number):
    if over_number <= 6:
        return 'Powerplay (1-6)'
    elif over_number <= 15:
        return 'Middle (7-15)'
    else:
        return 'Death (16-20)'
```

```python
df['over_phase']=df['over'].apply(get_phase)
```

```python
print('How many balls in each phase:')
print(df['over_phase'].value_counts())
```

```python
df['is_boundary']=df['batsman_runs'].apply(lambda x: 1 if x in [4, 6] else 0)
```

```python
print('Total boundaries hit in IPL history:',df['is_boundary'].sum())
```

```python
df['is_dot_ball']=((df['batsman_runs']==0)&(df['extras_type']=='none')).astype(int)
```

```python
print('Total dot balls in IPL history:',df['is_dot_ball'].sum())
```

```python
def run_cat(runs):
    if runs==0:
        return 'Dot'
    elif runs==1:
        return 'Single'
    elif runs==2:
        return'Double'
    elif runs==3:
        return'Triple'
    elif runs==4:
        return 'Four'
    else:
        return 'Six'
```

```python
df['run_category']=df['batsman_runs'].apply(run_cat)
```

```python
print('Count of each run category:')
print(df['run_category'].value_counts())
```

```python
print('Dataframe with new columns:')
print(df.columns.tolist())
```

```python
#Analysis Data
```

```python
team_total_runs = df.groupby('batting_team')['total_runs'].sum()
team_total_runs = team_total_runs.sort_values(ascending=False)
```

```python
print('Total runs scored by each team:')
print(team_total_runs)
```

```python
phase_analysis = df.groupby('over_phase')['total_runs'].mean()
```

```python
phase_order = ['Powerplay (1-6)','Middle (7-15)','Death (16-20)']
phase_analysis = phase_analysis.loc[phase_order]
phase_analysis = phase_analysis.round(3)
```

```python
print('Average runs scored per ball in each phase:')
print(phase_analysis)
print()
for phase, avg in phase_analysis.items():
    print(f'  {phase}: {avg} runs per ball on average')
```

```python
top_batters = df.groupby('batter')['batsman_runs'].sum()
top_batters = top_batters.sort_values(ascending=False)
top_batters = top_batters.head(10)
```

```python
print('Top 10 Run Scorers in IPL:')
rank = 1
for player_name,runs in top_batters.items():
    print(f'Rank {rank}:{player_name}—{runs} runs')
    rank=rank+1
```

```python
boundary_rows = df[df['is_boundary']==1]
```

```python
team_boundaries = boundary_rows.groupby('batting_team')['is_boundary'].sum()
team_boundaries = team_boundaries.sort_values(ascending=False)
```

```python
print('Total Boundaries (Fours + Sixes) by Team:')
print(team_boundaries)
```

```python
dot_ball=df.groupby('bowling_team').agg(
    Total_Balls=('ball','count'),
    Dot_Balls=('is_dot_ball','sum')
)
dot_ball['Dot_Ball_Percent']=(
    dot_ball['Dot_Balls']/dot_ball['Total_Balls']*100
).round(2)
```

```python
dot_ball=dot_ball.sort_values('Dot_Ball_Percent',ascending=False)
print('Dot Ball Percentage:')
print(dot_ball)
```
